```python
import pandas as pd
import matplotlib.pyplot as plt

# ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸÜ ŸÖŸÑŸÅ CSV
df = pd.read_csv(r"D:\ŸÖŸáÿßŸÖ ÿßÿ∂ÿßŸÅŸäÿ©\Patients Data\patients_data.csv")

# ÿπÿ±ÿ∂ ÿ£ŸàŸÑ 5 ÿµŸÅŸàŸÅ
print("üîç First 5 Patients:\n", df.head())

# ÿπÿØÿØ ÿßŸÑŸÖÿ±ÿ∂Ÿâ ÿßŸÑÿ•ÿ¨ŸÖÿßŸÑŸä
print("\nüë• Total Patients:", len(df))

# ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿπŸÜ ÿßŸÑÿ£ÿπŸÖÿßÿ±
print("\nüìä Age Statistics:\n", df['Age'].describe())

# ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ£ŸÖÿ±ÿßÿ∂
disease_counts = df['Disease'].value_counts()
print("\nü¶† Disease Distribution:\n", disease_counts)

# ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ (Outcome)
outcome_counts = df['Outcome'].value_counts()
print("\nüìà Outcome Distribution:\n", outcome_counts)

# ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿπŸÑÿßÿ¨ÿßÿ™
treatment_counts = df['Treatment'].value_counts()
print("\nüíä Treatment Types:\n", treatment_counts)

# ÿ±ÿ≥ŸÖ ÿ®ŸäÿßŸÜŸä ŸÑŸÑÿ£ŸÖÿ±ÿßÿ∂
plt.figure(figsize=(8, 5))
disease_counts.plot(kind='bar', color='skyblue', title='Disease Distribution')
plt.xlabel('Disease')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()

# ÿ±ÿ≥ŸÖ ÿ®ŸäÿßŸÜŸä ŸÑŸÑŸÜÿ™ÿßÿ¶ÿ¨
plt.figure(figsize=(6, 6))
outcome_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, title='Patient Outcomes')
plt.ylabel('')
plt.tight_layout()
plt.show()

# ÿ±ÿ≥ŸÖ ÿ®ŸäÿßŸÜŸä ŸÑŸÑÿπŸÑÿßÿ¨ÿßÿ™
plt.figure(figsize=(8, 5))
treatment_counts.plot(kind='barh', color='lightgreen', title='Treatment Types')
plt.xlabel('Number of Patients')
plt.ylabel('Treatment')
plt.tight_layout()
plt.show()
```
